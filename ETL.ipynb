```python
import pandas as pd
import yfinance as yf
```

```python
#Descargo data table wikipedia S&P 500
sp500url = "https://en.wikipedia.org/wiki/List_of_S%26P_500_companies"

data_table = pd.read_html(sp500url)
```

```python
#Hago una lista simbolos syp500
tickers = data_table[0]["Symbol"].tolist()
```

```python
data_table[0]
```

```python
#renombro para trabajar mas comodo
sector = data_table[0]
```

```python
#Sectores
sector["GICS Sector"].unique()
```

```python
sector
```

```python
#Hago listas en base a los sectores
listaInd=[]
listaHC=[]
listaInfTec=[]
listaComS=[]
listaConStap=[]
listaConDisc=[]
listaUtilities=[]
listaFinan=[]
listaMaterial=[]
listaRealEst=[]
listaEnergy=[]

for ind, a in enumerate(sector["GICS Sector"]):
    if a == "Industrials":
        listaInd.append(sector["Symbol"][ind])
    if a == "Health Care":
        listaHC.append(sector["Symbol"][ind])
    if a == "Information Technology":
        listaInfTec.append(sector["Symbol"][ind])
    if a == "Communication Services":
        listaComS.append(sector["Symbol"][ind])
    if a == "Consumer Staples":
        listaConStap.append(sector["Symbol"][ind])
    if a == "Consumer Discretionary":
        listaConDisc.append(sector["Symbol"][ind])
    if a == "Utilities":
        listaUtilities.append(sector["Symbol"][ind])
    if a == "Financials":
        listaFinan.append(sector["Symbol"][ind])
    if a == "Materials":
        listaMaterial.append(sector["Symbol"][ind])
    if a == "Real Estate":
        listaRealEst.append(sector["Symbol"][ind])
    if a == "Energy":
        listaEnergy.append(sector["Symbol"][ind])
```

```python
#Corroboro que ninguna industria haya quedado afuera
len(listaInd) + len(listaHC)+ len(listaInfTec)+ len(listaComS)+ len(listaConStap)+ len(listaConDisc)+ len(listaUtilities)+ len(listaFinan)+ len(listaMaterial)+ len(listaRealEst)+ len(listaEnergy)
```

```python
#Modifico Symbol para yfinance download en tickers
for i in range(len(tickers)):
    if tickers[i]== "BRK.B":
        tickers[i]= "BRK-B"
for i in range(len(tickers)):
    if tickers[i]=="BF.B":
        tickers[i]= "BF-B"
```

```python
#Modifico Symbol para yfinance download en listas=tickers
for i in range(len(listaFinan)):
    if listaFinan[i]== "BRK.B":
        listaFinan[i]= "BRK-B"
for i in range(len(listaConStap)):
    if listaConStap[i]=="BF.B":
        listaConStap[i]= "BF-B"
```

```python
#Descargo informacion Adj Close
data500AdjC = yf.download(tickers, start="2000-01-01",end="2023-01-03")["Adj Close"]
```

```python
#Descargo informacion Volume
data500Vol = yf.download(tickers, start="2000-01-01",end="2023-01-03")["Volume"]
```

```python
#I MAKE COLUMNS BY SECTORS,    ADJUST CLOSE  
data500AdjC["Total Industrial"] = data500AdjC[listaInd].sum(axis=1)
data500AdjC["Total Health Care"] = data500AdjC[listaHC].sum(axis=1)
data500AdjC["Total Information Technology"] = data500AdjC[listaInfTec].sum(axis=1)
data500AdjC["Total Communication Services"] = data500AdjC[listaComS].sum(axis=1)
data500AdjC["Total Consumer Staples"] = data500AdjC[listaConStap].sum(axis=1)
data500AdjC["Total Consumer Discretionary"] = data500AdjC[listaConDisc].sum(axis=1)
data500AdjC["Total Utilities"] = data500AdjC[listaUtilities].sum(axis=1)
data500AdjC["Total Financials"] = data500AdjC[listaFinan].sum(axis=1)
data500AdjC["Total Materials"] = data500AdjC[listaMaterial].sum(axis=1)
data500AdjC["Total Real Estate"] = data500AdjC[listaRealEst].sum(axis=1)
data500AdjC["Total Energy"] = data500AdjC[listaEnergy].sum(axis=1)

#I MAKE COLUMNS BY SECTORS, VOLUME

data500Vol["Total Industrial"] = data500Vol[listaInd].sum(axis=1)
data500Vol["Total Health Care"] = data500Vol[listaHC].sum(axis=1)
data500Vol["Total Information Technology"] = data500Vol[listaInfTec].sum(axis=1)
data500Vol["Total Communication Services"] = data500Vol[listaComS].sum(axis=1)
data500Vol["Total Consumer Staples"] = data500Vol[listaConStap].sum(axis=1)
data500Vol["Total Consumer Discretionary"] = data500Vol[listaConDisc].sum(axis=1)
data500Vol["Total Utilities"] = data500Vol[listaUtilities].sum(axis=1)
data500Vol["Total Financials"] = data500Vol[listaFinan].sum(axis=1)
data500Vol["Total Materials"] = data500Vol[listaMaterial].sum(axis=1)
data500Vol["Total Real Estate"] = data500Vol[listaRealEst].sum(axis=1)
data500Vol["Total Energy"] = data500Vol[listaEnergy].sum(axis=1)


```

```python
#Hago lista de sectores para filtrar por columnas


listasectores = ["Total Industrial","Total Health Care","Total Information Technology","Total Communication Services","Total Consumer Staples","Total Consumer Discretionary",
                "Total Utilities","Total Financials","Total Materials","Total Real Estate","Total Energy"]
```

```python
#Podemos calcular Volatilidad, lo hacemos con el Sector Industral
VarianzaIndustrial= data500AdjC["Total Industrial"]
```

```python
VarianzaIndustrial.index = pd.to_datetime(VarianzaIndustrial.index)
```

```python
df_mes = VarianzaIndustrial.resample('M').mean()
```

```python
df_retornos = df_mes.pct_change()

volatilidad = df_retornos.std()

volatilidad
```
